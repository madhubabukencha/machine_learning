### <span style="color:#3C4048; font-weight: bold; font-size: 18px; font-family: Gill Sans, sans-serif;">Importing Necessary Packages</span>

```python
import matplotlib.pyplot as plt
# doc: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
%matplotlib inline
```

### <span style="color:#3C4048; font-weight: bold; font-size: 18px; font-family: Gill Sans, sans-serif;">Generating Data</span>

```python
data = make_blobs(n_samples=100,  # Number of samples
                  n_features=2,  # Each sample has 2 features
                  centers=4,
                  cluster_std=1.8,
                  random_state=101
                  )
```

```python
data
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1],
            c=data[1], # color for each sample
            cmap="rainbow")
```

### <span style="color:#3C4048; font-weight: bold; font-size: 18px; font-family: Gill Sans, sans-serif;">Building The Model</span>

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
# It has same number of dimentions of that your dataset features
kmeans.cluster_centers_
```

```python
# predicted label 
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6))

ax1.set_title("K Means")
ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_, cmap="rainbow")

ax2.set_title("Original")
ax2.scatter(data[0][:,0], data[0][:, 1], c=data[1], cmap="rainbow")
```
