<font size=5 color="color:#290001;">Notations:</font>
<ul>
    <li><b>m</b>: Number of training examples(Number of rows in the dataset)</li>
    <li><b>x's</b>: "Input" variable or Features or Independent Variable</li>
    <li><b>y's</b>:  "Output" variable or Target variable or Dependent Variable</li>
    <li><b>(x, y)</b>: Represents a single training example in the dataset</li>
    <li><b>(x<sup>(i)</sup>, y<sup>(i)</sup>)</b>: a list of m training examples; i=1,...,m—is called a training set
    <li>We will also use X to denote the space of input values, and Y to denote the space of output values. In this example, X = Y = ℝ.</li>
</ul>

### <font size=5 color="color:#290001;">How Supervised Learning algorithms works:</font>

<p>To describe the supervised learning problem slightly more formally, our goal is, given a training set, to learn a function h : X → Y so that h(x) is a “good” predictor for the corresponding value of y. For historical reasons, this function h is called a hypothesis.</p>
<img src="Images\model_representation.png" alt="model_representation" style="width: 400px;"/>
<p>When the target variable that we are trying predict is continous, such housing price, stack, electricity bill etc.., We call the learning problem a regression problem. When y can take on only a small number of discrete values (such as if, given the living area, we wanted to predict if a dwelling is a house or an apartment, say), we call it a classification problem</p>
<br><br>

<b>h</b> is an hypothesis function. It is represented as <b>h<sub>θ</sub>(x) = θ<sub>0</sub>+θ<sub>1</sub>x</b> <br>
<p> This is called <b>Linear Regression with one Variable or Univariate Linear Regression</b></p>

## Basic Linear Example Graph

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

x = np.array([100, 25, 50, 200, 30, 45, 23, 70, 120, 135, 170, 79, 80])
y = np.array([8, 7, 13, 20, 3, 9, 15, 7, 12, 18, 9, 6, 4])

plt.plot(x, y, 'x')
plt.title("Linear Regression Example")
plt.xlabel("Cofee Rates")
plt.ylabel("Tips")
plt.rcParams["figure.figsize"] = (10, 10)
plt.grid(True)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, label='asdfs')
plt.show()
```

In the above graph straight line represent the hyphothesis
